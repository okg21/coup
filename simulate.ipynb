```python
NUM_GAMES = 10000

def run_simulation(players, shuffle_turn_order=True, num_games=NUM_GAMES, turn_limit=50):
    winners = []
    for _ in range(num_games):
        if shuffle_turn_order:
            random.shuffle(players)
        game = Game(players)
        turn_count = 0
        while len(game.game_state['players']) > 1 and turn_count < turn_limit:
            game.simulate_turn()
            turn_count += 1
        if turn_count == turn_limit:
            winner = 'draw'
        else:
            winner = game.game_state['players'][0].name
        winners.append(winner)
    return winners

def plot_winners(winners, player_types, title):
    player_type_wins = {p_type : 0 for p_type in player_types.values()}
    for p_type in player_type_wins.keys():
        player_type_wins[p_type] = len([winner for winner in winners if player_types[winner] == p_type])
    plt.figure()
    p1 = plt.bar(player_type_wins.keys(), player_type_wins.values(), color='forestgreen')
    for rect1 in p1:
        percent_wins = round(rect1.get_height() / len(winners) * 100, 4)
        height = rect1.get_height()
        plt.annotate( "{}%".format(percent_wins),(rect1.get_x() + rect1.get_width()/2, height / 2),ha="center",va="bottom",fontsize=15)
    plt.title(title)
    plt.show()
```

```python
players = [Player('random_player', RANDOM_FUNCS), Player('truth_player', TRUTH_FUNCS)]
player_types = {'random_player': 'random', 'truth_player': 'truth'}
winners = run_simulation(players, turn_limit=1000)
plot_winners(winners, player_types, "random v.s. truth wins")
```

```python
def plot_lies(agent):
    """
    Display the total lie% and the relative lie% for each action.
    """
    actions = agent.list_of_actions
    lies = agent.did_action_lie
    lies = [lie for idx, lie in enumerate(lies) if actions[idx][2] != 'Income' and actions[idx][2] != 'Foreign Aid' and actions[idx][2] != 'Coup']

    total_lie_freq = round(sum(lies) / len(lies), 2)

    action_types = set([a[2] for a in actions])
    type_lies = {}
    type_lie_freq = {'Overall': 0, 'Tax': 0, 'Exchange': 0, 'Steal': 0, 'Assassinate': 0}
    type_lie_freq['Overall'] = total_lie_freq
    for type in action_types:
        type_lies[type] = [lie for idx, lie in enumerate(lies) if actions[idx][2] == type]
        if type != 'Income' and type != 'Foreign Aid' and type != 'Coup' and len(type_lies[type]) > 0:
            type_lie_freq[type] = round(sum(type_lies[type]) / len(type_lies[type]), 2)
    plt.figure()
    p1 = plt.bar(type_lie_freq.keys(), type_lie_freq.values(), color=['indianred', 'plum', 'lightgreen', 'cornflowerblue', 'lightslategray'])
    for rect1 in p1:
        height = rect1.get_height()
        plt.annotate( "{}%".format(rect1.get_height()),(rect1.get_x() + rect1.get_width()/2, height / 2),ha="center",va="bottom",fontsize=15)
    plt.title('Lie frequency of {0}'.format(agent.name))
    plt.show()


def plot_action_types(agent):
    """
    Display the frequency of actions taken by the agent.
    """
    actions = agent.list_of_actions
    action_types = set([a[2] for a in actions])
    action_freq = {'Tax': 0, 'Exchange': 0, 'Income': 0, 'Assassinate': 0, 'Steal': 0, 'Foreign Aid': 0, 'Coup': 0}
    for type in action_types:
        action_freq[type] = round(len([a for a in actions if a[2] == type]) / len(actions), 3)
    fig = plt.figure()
    fig.set_figwidth(10)
    p1 = plt.bar(action_freq.keys(), action_freq.values(), color=['plum', 'lightgreen', 'gainsboro', 'cornflowerblue', 'lightslategray', 'gainsboro', 'gainsboro'])
    for rect1 in p1:
        height = rect1.get_height()
        plt.annotate( "{}%".format(rect1.get_height()),(rect1.get_x() + rect1.get_width()/2, height / 2),ha="center",va="bottom",fontsize=15)
    plt.title('Action frequency of {0}'.format(agent.name))
    plt.show()
```
